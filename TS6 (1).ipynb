```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("daily-min-temperatures.csv",parse_dates=['Date'],index_col='Date')
df.head()
```

```python
df.plot(title="Daily Min temp",figsize=(14,8),legend=None)
plt.xlabel("Date")
plt.ylabel("Temp")
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
fig,axes = plt.subplots(2,1,figsize=(10,8))

plot_acf(df['Temp'],lags=30,ax=axes[0],title="Autocorrelation")
plot_pacf(df['Temp'],lags=30,ax=axes[1],title="Partial Autocorrelation")
plt.tight_layout()
plt.show()
```
