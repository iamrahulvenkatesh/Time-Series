```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
custom_col = ['WSR0','WSR1']
df = pd.read_csv("eighthr.csv",header=None,parse_dates=[0],index_col=[0],na_values=["?"],names = custom_col+list(range(3,74)))
df.head()
```

```python
df = df.dropna()
df
```

```python
df['WSR0'] = df['WSR0'].astype(float)
df
```

```python
fig,axes = plt.subplots(2,1,figsize=(14,8))
df['WSR1'].plot(ax=axes[0],title="WSR1")
df['WSR0'].plot(ax=axes[1],title="WSR0")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['WSR0'])
print(result)
if result[1]<0.05:
    print('stationary')
else:
    print('not')
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['WSR1'])
print(result)
if result[1]<0.05:
    print('stationary')
else:
    print('not')
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.statespace.varmax import VARMAX
```

```python
print("WSR0 causes WSR1")
granger1 = grangercausalitytests(df[['WSR0','WSR1']],2)
print("WSR1 causes WSR0")
granger1 = grangercausalitytests(df[['WSR1','WSR0']],2)
```
