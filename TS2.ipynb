```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("shampoo_sales.csv")
df.head()
```

```python
df.set_index('Month',inplace=True)
```

```python
df.head()
```

```python
plt.hist(df['Sales'],bins=8,edgecolor='black')
```

```python
plt.plot(df.index,df['Sales'])
plt.xticks(rotation=90)
```

```python
df.index=pd.to_datetime(df.index,format='%d-%m')
df.index = df.index.map(lambda t:t.replace(year=2023))
```

```python
df['Year']=df.index.year
df['Month']=df.index.month
df['Day']=df.index.day
df
```

```python
df['lag_1']=df['Sales'].shift(1)
df
```

```python
df['lag_2']=df['Sales'].shift(2)
df
```

```python
df['rolling_mean']=df['Sales'].rolling(5).mean()
df['rolling_std']=df['Sales'].rolling(5).std()
df['rolling_min']=df['Sales'].rolling(5).min()
df['rolling_max']=df['Sales'].rolling(5).max()
df
```

```python
df['expanding_mean']=df['Sales'].expanding().mean()
df['expanding_std']=df['Sales'].expanding().std()
df['expanding_min']=df['Sales'].expanding().min()
df['expanding_max']=df['Sales'].expanding().max()
df
```
